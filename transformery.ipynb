Użycie wielojęzycznego modelu tłumaczeniowego

```python
from transformers import MBartForConditionalGeneration, MBart50Tokenizer
# Inicjalizacja modelu i tokenizera
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50Tokenizer.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

# Określenie języka docelowego (polski)
tokenizer.src_lang = "en_XX"
```

```python
text = "Each value is associated with a key and is used to construct the output of the attention layer. When a query and a key match well, which basically means that they have a high attention score, the corresponding value is emphasized in the output."

# Tokenizacja
inputs = tokenizer(text, return_tensors="pt")

# Tłumaczenie na polski
translated_tokens = model.generate(inputs.input_ids, forced_bos_token_id=tokenizer.lang_code_to_id["pl_PL"])
translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

print("Przetłumaczony tekst:", translated_text)
```

3. Streszczanie tekstu (po polsku)

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Ładowanie modelu i tokenizera dla T5
model_name = "allegro/plt5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Tekst do streszczenia
text = """
Hugging Face to platforma, która umożliwia łatwe korzystanie z modeli głębokiego uczenia dla języka naturalnego.
Oferuje gotowe modele dla tłumaczeń, streszczeń, analizy sentymentu i wielu innych zastosowań.
Wspiera rozwój społeczności open-source, dzięki czemu użytkownicy mogą dzielić się swoimi modelami i pomysłami.
"""

# Przygotowanie tekstu dla modelu
input_text = f"streść: {text}"
inputs = tokenizer(input_text, return_tensors="pt", truncation=True)

# Generowanie streszczenia
outputs = model.generate(inputs.input_ids, max_length=50, num_beams=4, early_stopping=True)
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Streszczenie:", summary)




```

4. Odpowiadanie na pytania (w języku polskim)

```python
from transformers import pipeline

# Pipeline do odpowiadania na pytania
question_answerer = pipeline("question-answering", model="allegro/herbert-base-cased")

# Kontekst i pytanie
context = "Hugging Face oferuje biblioteki do analizy języka naturalnego, tłumaczeń, streszczeń i wielu innych zadań."
question = "Co oferuje język?"

# Generowanie odpowiedzi
result = question_answerer(question=question, context=context)
print("Odpowiedź:", result['answer'])
```

6. Klasyfikacja tematów tekstu w języku polskim


```python
from transformers import pipeline

# Pipeline do klasyfikacji zero-shot
text_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Tekst do analizy
text = "Uwielbiam uczyć się nowych technologii i sztucznej inteligencji."

# Kategorie do klasyfikacji
labels = ["AI", "sport", "nauka", "edukacja"]

# Klasyfikacja
result = text_classifier(text, candidate_labels=labels)

print("Wynik klasyfikacji:", result)
```
