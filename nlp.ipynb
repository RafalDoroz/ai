```python
#### 3. Podstawowe użycie pipeline do analizy sentymentu:

from transformers import pipeline

# Utworzenie pipeline do analizy sentymentu
sentiment_analyzer = pipeline("sentiment-analysis")

# Analiza przykładowych tekstów
texts = [
    "I love programming! It's so exciting!",
    "This weather is terrible. I feel so gloomy."
]

results = sentiment_analyzer(texts)

# Wyświetlenie wyników
for text, result in zip(texts, results):
    print(f"Tekst: {text}")
    print(f"Emocja: {result['label']}, Pewność: {result['score']:.2f}")
```

```python
#### 4. Tłumaczenie tekstu

translator = pipeline("translation_en_to_fr")

text_to_translate = "I love working with Hugging Face!"
translation = translator(text_to_translate)

print("Tłumaczenie:", translation[0]['translation_text'])
```

```python
#### 5. Generowanie tekstu

text_generator = pipeline("text-generation", model="gpt2")

prompt = "Once upon a time in a distant galaxy"
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)

print("Wygenerowany tekst:")
print(generated_text[0]['generated_text'])
```

```python
#### 6. Rozpoznawanie nazwanych jednostek (NER)  Możesz rozpoznawać nazwane jednostki w tekście, np. osoby, lokalizacje czy organizacje:

ner = pipeline("ner", grouped_entities=True)

text = "Barack Obama was the 44th President of the United States and lived in Washington D.C."
entities = ner(text)

print("Rozpoznane jednostki:")
for entity in entities:
    print(f"Tekst: {entity['word']}, Typ: {entity['entity_group']}, Pewność: {entity['score']:.2f}")
```

```python
#### 7. Lematyzacja  Hugging Face nie oferuje bezpośredniej funkcji lematyzacji w pipeline, możemy użyć modelu przeszkolonego do tokenizacji i przetwarzania tekstu. Alternatywnie możemy skorzystać z biblioteki **spaCy**, która doskonale radzi sobie z lematyzacją:


import spacy

# Wczytanie modelu spaCy do języka angielskiego
nlp = spacy.load("en_core_web_sm")

text = "The children are playing in the garden."
doc = nlp(text)

print("Lematyzacja słów:")
for token in doc:
    print(f"Słowo: {token.text}, Lemma: {token.lemma_}")
```

```python
#### 8. Tworzenie n-gramów


from nltk.util import ngrams

# Przykładowy tekst
tokenized_text = "The quick brown fox jumps over the lazy dog".split()

# Generowanie bigramów (2-gramów)
bigrams = list(ngrams(tokenized_text, 2))

# Generowanie trigramów (3-gramów)
trigrams = list(ngrams(tokenized_text, 3))

print("Bigramy:", bigrams)
print("Trigramy:", trigrams)
```

```python
#### 9. Klasyfikacja tekstu


from transformers import pipeline

# Pipeline do klasyfikacji tekstu
text_classifier = pipeline("text-classification")

texts = [
    "The movie was fantastic! I really enjoyed it.",
    "The food was cold and tasteless."
]

results = text_classifier(texts)

print("Wyniki klasyfikacji:")
for text, result in zip(texts, results):
    print(f"Tekst: {text}")
    print(f"Kategoria: {result['label']}, Pewność: {result['score']:.2f}")
```

```python
#### 10. Wektoryzacja tekstu i klasyfikacja

from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Model do osadzeń
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Funkcja do generowania osadzeń
def get_embeddings(texts):
    tokens = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        embeddings = model(**tokens).last_hidden_state.mean(dim=1)
    return embeddings.numpy()

# Dane wejściowe
texts = [
    "I love programming!",
    "The weather is gloomy.",
    "Python is amazing!",
    "I feel sad today.",
    "Machine learning is fascinating!",
    "I'm not feeling well."
]
labels = [1, 0, 1, 0, 1, 0]  # 1 - pozytywne, 0 - negatywne

# Podział na zbiory treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# Generowanie osadzeń
X_train_embeddings = get_embeddings(X_train)
X_test_embeddings = get_embeddings(X_test)

# Trening klasyfikatora
classifier = SVC()
classifier.fit(X_train_embeddings, y_train)

# Predykcja i ocena
predictions = classifier.predict(X_test_embeddings)
accuracy = accuracy_score(y_test, predictions)

print(f"Dokładność klasyfikacji: {accuracy:.2f}")

# Klasyfikacja własnego zdania
own_text = input("Podaj własne zdanie do klasyfikacji: ")
own_embedding = get_embeddings([own_text])
own_prediction = classifier.predict(own_embedding)
category = "Pozytywne" if own_prediction[0] == 0 else "Negatywne"
print(f"Twój tekst: {own_text}")
print(f"Kategoria: {category}")
```
