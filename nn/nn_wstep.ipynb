```python
from scipy.special import expit
import matplotlib.pyplot as plt
import numpy as np
#matplotlib inline



# sigmoid activation function
def sigmoid(z):
  return 1.0 / (1 + np.exp(-z))

# tanh activation function
def tanh(z):
  return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))


#  ReLU activation function
def relu(z):
    return np.maximum(0, z)

#  leaky_relu activation function
def leaky_relu(z, alpha=0.01):
  return np.where(z > 0, z, z * alpha)


# Softmax activation function
def softmax(z):
    exp_values = np.exp(z)  # Stabilizacja numeryczna
    #exp_values = np.exp(z - np.max(z))  # Stabilizacja numeryczna
    return exp_values / np.sum(exp_values)

x = np.linspace(-2,2)

w1 = sigmoid(np.dot(x, -6.75) + 3.42) * 4.69 + 8.11   #blue
w2 = sigmoid(np.dot(x, 6.75) - 4.42) * 7.69 + 8.11   #orange

plt.plot(x,w1)
plt.plot(x,w2)
plt.grid()
plt.show()


# s1 = sigmoid(np.dot(x, -0.75) + 1.42) * 4.69 + 8.11   #blue
# s2 = sigmoid(np.dot(x, 13.91) + 8.34) * 11.69 - 8.11   # orange
# s3 = sigmoid(np.dot(x, -22.03) - 1.44) * 22.69 + 1.11  # green
# ss = s1 + s2 + s3 - 11                                 # red

# plt.plot(x,s1)
# plt.plot(x,s2)
# plt.plot(x,s3)
# plt.plot(x,ss)
# plt.grid()
# plt.show()



```
